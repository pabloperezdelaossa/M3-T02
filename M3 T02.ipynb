### - Exercici 2.c

Fes un informe complet del dataset:

    Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).

```python
# faig una columna que indica si ha hagut retras, independentment del tipus

import numpy as np

def retras_si_no(row):  
    if row['CarrierDelay'] > 0 or row['WeatherDelay']  > 0 or row['NASDelay']  > 0 or row['SecurityDelay']  > 0 or row['LateAircraftDelay']  > 0:
        return 'Retras'
    else:
        pass
    return 'A temps'

vols2['Delayed'] = vols2.apply(lambda row: retras_si_no(row), axis=1)


# dona un avis, pero la columna ha sigut creada
```

```python
# Fem una nova columna amb la velocitat del vol
vols2['Speed'] = vols2.apply(lambda row: row.Distance / row.ActualElapsedTime, axis = 1)

vols2[:10]

# de nou, dona un avís però la columna ha estat creada
```

```python
# aixo ens dona el total de vols per cada companyia, només amb retras
def nomes_retras():
    return vols2.groupby('UniqueCarrier')['Delayed'].apply(lambda x: x[x == 'Retras'].count())

nomes_retras().sort_values(ascending=False)
```

```python
# aixo ens dona el percentatge de vols amb retras per cada companyia. Faig servir la mateixa estructura de codi d'abans
# vols amb retras *100 / (vols amb retras + vols a temps)
def operador_retras():
    
    return vols2.groupby('UniqueCarrier')['Delayed'].apply(lambda x: x[x == 'Retras'].count()) * 100 / (( vols2.groupby('UniqueCarrier')['Delayed'].apply(lambda x: x[x == 'Retras'].count())
  )+(vols2.groupby('UniqueCarrier')['Delayed'].apply(lambda x: x[x == 'A temps'].count())
 ))

operador_retras().sort_values(ascending=False)
```

### - Exercici 2.e

Fes un informe complet del dataset:

    Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards.


```python
# Els vols mes llargs (en temps). Només selecciono unes poques columnes
def temps_vol():
    vols_llargs = vols2[['UniqueCarrier','Origin','Dest','FlightNum', 'ActualElapsedTime']]
    return vols_llargs.groupby(['FlightNum','UniqueCarrier','Origin','Dest']).mean('ActualElapsedTime')

temps_vol().sort_values(by="ActualElapsedTime", ascending=False)[:10]
```

```python
# Els vols mes llargs (en temps). Només selecciono unes poques columnes
def distancia_vol():
    vols_llargs = vols2[['UniqueCarrier','Origin','Dest','FlightNum', 'Distance']]
    return vols_llargs.groupby(['FlightNum','UniqueCarrier','Origin','Dest']).mean('Distance')

distancia_vol().sort_values(['Distance'], ascending=False)[:10]
```

```python
# Fem una nova columna amb el retard total (suma dels 5 tipus de retards)
vols2['Tot_delay'] = vols2.apply(lambda row: row.CarrierDelay + row.WeatherDelay + row.NASDelay + row.SecurityDelay + row.LateAircraftDelay, axis = 1)

vols2[['CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay','Tot_delay']].head()

# de nou, dona un avís però la columna ha estat creada
```

```python
# Els vols mes endarrerits. Només selecciono unes poques columnes
def temps_vol():
    vols_llargs = vols2[['UniqueCarrier','Origin','Dest','FlightNum', 'Tot_delay']]
    return vols_llargs.groupby(['FlightNum','UniqueCarrier','Origin','Dest']).max('Tot_delay')

temps_vol().sort_values(by="Tot_delay", ascending=False)[:10]

# Aquest codi mostra el vol que han patit mes retards per cada ruta
```

```python
vols2.sort_values(by="Tot_delay",ascending=False)[:10]

# Aquest codi mostra els vols que han patit més retard en general, però el resultat es una mica brut, 
# ja que surten totes les columnes i el retard total surt a la dreta de tot.
```

```python
# Els vols mes endarrerits. Només selecciono unes poques columnes
def retard_vol():
    vols_retard = vols2[['UniqueCarrier','Origin','Dest','FlightNum', 'Tot_delay']]
    return vols_retard

retard_vol().sort_values(by="Tot_delay", ascending=False)[:10]

# Aquest codi SI mostra els vols que han patit mes retards en global
```

### Retards a la sortida

```python
# Aeroports amb més retards de sortida

def retard_sortida():
    vols_retard = vols2[['Origin','ArrDelay']]
    return vols_retard.groupby(['Origin']).mean('ArrDelay')

retard_sortida().sort_values(by="ArrDelay", ascending=False)[:10]
```

#### Comentari: 
Al codi de dalt hi veiem quin son els aeroports amb mes retard promig.
Al codi de abix hi afegim la columna de la companyia (UniqueCarrier) i obtenim el retard promig per aeroport i companyia

```python
# Aeroports amb més retard de sortida per companyia aerea

def retard_sortida():
    vols_retard = vols2[['Origin','UniqueCarrier','ArrDelay']]
    return vols_retard.groupby(['Origin','UniqueCarrier']).mean('ArrDelay')

retard_sortida().sort_values(by="ArrDelay", ascending=False)[:10]
```

#### Qué passa amb la companyia NW?

Em crida l'atenció que els temps de rtards de NW son molt alts. Surt 5 cops a la llista de dalt. Als dos propers codis coparem els retards entre la companyia NW i AQ, seleccionant nomès una de les dues companyies en cada codi 

```python
# Aeroports amb més retard de sortida per companyia aerea: NOMES COMPANYIA NW

def retard_sortida_NW():
    
    vols_retard = vols2[['Origin','UniqueCarrier','ArrDelay']]
    vols_retard2 = vols_retard.loc[vols_retard['UniqueCarrier'] == 'NW']
    return vols_retard2.groupby(['Origin','UniqueCarrier']).mean('ArrDelay')

retard_sortida_NW().sort_values(by="ArrDelay", ascending=False)[:10]
```

```python
# Aeroports amb més retard de sortida per companyia aerea: NOMES COMPANYIA AQ

def retard_sortida_AQ():
    
    vols_retard = vols2[['Origin','UniqueCarrier','ArrDelay']]
    vols_retard2 = vols_retard.loc[vols_retard['UniqueCarrier'] == 'AQ']
    return vols_retard2.groupby(['Origin','UniqueCarrier']).mean('ArrDelay')

retard_sortida_AQ().sort_values(by="ArrDelay", ascending=False)[:10]
```

#### Conclusió preliminar

La companyia NW te uns retards de sortida molt elevats comparats amb AQ
